```python
import pandas as pd
import numpy as np
import tensorflow as tf
import sklearn

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import utils as np_utils
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
```

```python
#Carregando o dataset

base = pd.read_csv('/content/iris.csv')

#Visualizando as primeiras linhas do dataset
print(base.head())

#Visualizando as informações do dataset
print(base.info())
```

```python
#Separando as features e as classes
X = base.iloc[:, 0:4].values
X
y = base.iloc[:, 4].values
y

#Convertendo as classes para valores numéricos
labelencoder = LabelEncoder()
y = labelencoder.fit_transform(y)
y = np_utils.to_categorical(y)
y

```

```python
#Como temos 3 classes, vvamos utilizar soft-max de função de ativação
#Temos 4 neuronios de entrada e 3 de saida
X_treinamento, X_teste, y_treinamento, y_teste = train_test_split(X, y, test_size=0.25)
X_treinamento.shape, X_teste.shape
```

```python
y_treinamento.shape, y_teste.shape
```

```python
rede_neural = tf.keras.Sequential([
    tf.keras.layers.InputLayer(input_shape=(4,)),
    tf.keras.layers.Dense(units=4, activation='relu'),
    tf.keras.layers.Dense(units=4, activation='relu'),
    tf.keras.layers.Dense(units=3, activation='softmax')
])
```

```python
rede_neural.summary()
```

```python
rede_neural.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['categorical_accuracy'])
```

```python
rede_neural.fit(X_treinamento, y_treinamento, batch_size=10, epochs=1000)
```

```python
rede_neural.evaluate(X_teste, y_teste)
```

```python
previsoes = rede_neural.predict(X_teste)
previsoes
```

```python
previsoes = previsoes > 0.5
previsoes
```

```python
y_teste2 = [np.argmax(t) for t in y_teste]
previsoes2 = [np.argmax(t) for t in previsoes]
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_teste2, previsoes2)
```

```python
confusion_matrix(y_teste2, previsoes2)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from ipywidgets import interact, widgets

sns.set(style="whitegrid")

class_names = ['Setosa', 'Versicolor', 'Virginica']

def predict_and_visualize(sepal_length, sepal_width, petal_length, petal_width):
    new_observation = np.array([[sepal_length, sepal_width, petal_length, petal_width]])
    prediction = rede_neural.predict(new_observation)
    predicted_class = np.argmax(prediction)
    predicted_class_name = class_names[predicted_class]

    print(f"\033[1mPredicted class:\033[0m {predicted_class_name} (index: {predicted_class})")
    print(f"Prediction probabilities: {prediction[0]}")

    # Plot
    plt.figure(figsize=(8, 5))
    bars = sns.barplot(x=class_names, y=prediction[0], palette='Set2')
    plt.title("Probability per Iris Species", fontsize=14)
    plt.xlabel("Species")
    plt.ylabel("Probability")
    plt.ylim(0, 1)

    # Add labels on top of each bar
    for i, p in enumerate(prediction[0]):
        bars.text(i, p + 0.02, f"{p:.2f}", ha='center', fontsize=12, weight='bold')

    plt.tight_layout()
    plt.show()

# Sliders (sem mudanças)
sepal_length_slider = widgets.FloatSlider(min=4, max=8, step=0.1, value=5.8, description="Sepal Length:")
sepal_width_slider = widgets.FloatSlider(min=2, max=4.5, step=0.1, value=3, description="Sepal Width:")
petal_length_slider = widgets.FloatSlider(min=1, max=7, step=0.1, value=4, description="Petal Length:")
petal_width_slider = widgets.FloatSlider(min=0.1, max=2.5, step=0.1, value=1.2, description="Petal Width:")

interact(predict_and_visualize,
         sepal_length=sepal_length_slider,
         sepal_width=sepal_width_slider,
         petal_length=petal_length_slider,
         petal_width=petal_width_slider);
```

```python
import gradio as gr

#rede_neural = keras.models.load_model('modelo_iris.h5')

class_names = ['Setosa', 'Versicolor', 'Virginica']

def predict_iris(sepal_length, sepal_width, petal_length, petal_width):
    new_observation = np.array([[sepal_length, sepal_width, petal_length, petal_width]])
    prediction = rede_neural.predict(new_observation)
    return {class_names[i]: float(prediction[0][i]) for i in range(3)}

# Interface Gradio
iface = gr.Interface(
    fn=predict_iris,
    inputs=[
        gr.Slider(4, 8, value=5.8, label="Sepal Length"),
        gr.Slider(2, 4.5, value=3.0, label="Sepal Width"),
        gr.Slider(1, 7, value=4.0, label="Petal Length"),
        gr.Slider(0.1, 2.5, value=1.2, label="Petal Width"),
    ],
    outputs=gr.Label(num_top_classes=3),
    title="🌸🥀 Deep learning para previsão: Iris",
    description="Use os controles para prever a espécie de uma flor Iris com uma rede neural treinada."
)

# Executar localmente
iface.launch(inbrowser=True)




```
